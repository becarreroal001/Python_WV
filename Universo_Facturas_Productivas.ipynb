### Consiguir el universo de facturas productivas completo (desde marzo del 2014) 

Los recuperos pueden realizarce con facturas de hasta 48 meses atras (4 aÃ±os)

```python
import pandas
```

```python
# Cabecera
cab = ['SUFIJO','PREFIJO','PRECIO_NET','PERIODO_DD','NRO_VIAJE','NRO_PROV','NRO_DESPAC','NRO727','MANUAL','IVA_FACT',
       'IMPIVA_UNI','FACPROV_NS','FACPROV_NN','FACPROV_MM','FACPROV_DD','FACPROV_AA','DESCRIPCIO','DENOMIN_PR',
       'CUIT_PROV','COD_DOC','CANTIDAD','CANT_ASIGN','BASICO']
```

```python
factura = pandas.read_csv(r'C:\Volkswagen\facturas\facturas_ene18.csv', encoding = 'unicode_escape', 
                          header= 1, names = cab, low_memory=False, 
                          converters={'FACPROV_NN': lambda x: str(x), 'FACPROV_NS': lambda x: str(x),
                                      'COD_DOC': lambda x: str(x), 'CUIT_PROV': lambda x: str(x)})
# Concatenar las columnas fecha y numero de factura
factura['FECHA_FACT'] = factura['FACPROV_DD'].astype(str)+'/'+factura['FACPROV_MM'].astype(str)+'/'+factura['FACPROV_AA'].astype(str)
factura['NRO_FACT']= factura['FACPROV_NN']+factura['FACPROV_NS']
factura['CLAVE']= factura['NRO_FACT']+factura['CUIT_PROV']+factura['COD_DOC']
factura = factura[['CLAVE','FECHA_FACT','COD_DOC','PERIODO_DD','CUIT_PROV','NRO_PROV','DENOMIN_PR','NRO_FACT',
                               'PRECIO_NET','IVA_FACT','DESCRIPCIO','BASICO','PREFIJO','SUFIJO','CANTIDAD',
                               'CANT_ASIGN','IMPIVA_UNI']]
```

```python
febrero20 = factura.sort_values('PERIODO_DD') # Tengo facturas desde marzo del 2016
```

```python
enero18 = factura.sort_values('PERIODO_DD') # Tengo facturas desde marzo del 2014
```

```python
fac = febrero20[febrero20['PERIODO_DD'] >= 201801]
```

```python
facturas_prod = pandas.concat([enero18,fac], axis = 0).reset_index()
```

```python
facturas_prod = facturas_prod.drop(facturas_prod[facturas_prod['PERIODO_DD']== 0].index)
```

```python
facturas_prod['CANT_ASIGN'] = 0
```
